```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
img_path = '/content/drive/MyDrive/baboon.BMP'
img = cv2.imread(img_path)
```

```python
'''
To turn an image into its negative, we just need to subtract each pixel of the image from the maximum pixel
In the below, we have tried the above way to see the result of the "Baboon" image as a reference image
'''

img_negative = 255 - img

fig, ax = plt.subplots(ncols = 2 , nrows = 1 ,figsize=(16 ,16))
ax[0].imshow(img[:,:,::-1])
ax[0].set_title("Actuall Image")
ax[0].axis('off')

ax[1].imshow(img_negative[:,:,::-1])
ax[1].set_title("Negative Image")
ax[1].axis('off')
```
